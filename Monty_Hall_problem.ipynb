# Monty Hall problem

*   Let x be the door with the car
*   Player makes a choice
    * if he chose x:
        * Host will open y,z
    * if he choses y or z:
      * host will open either not y (i.e. z) or not z(i.e. y)

* Player has several strategies:
    * He can switch doors
        * if he chose x:
        Host opened either y or z(let's call the door he did not open j)
          * He will switch to j
    * He can stay with his choice
    * Choose randomly




##Code outline
* Generate a set wich contains the doors
* Randomly pick the door with the car (call it x)
* Randomly select a door(Player's choice call it p)
* Given the choice, the host will open a door(if p= x... else...)
* store in a variable the door opened(opened_door)

* simulate each of the strategies --> store the number of times the final choice of the player is x(call it success and store the number of successes/number of trials)

```python
doors = np.array(['x','y','z'])
door_car = np.random.choice(doors)
player_choice = np.random.choice(doors)
car_index = np.where(doors == door_car)[0][0] #the two [0] are to get the first coincidence only
mask = np.ones(len(doors), dtype = bool)
mask[car_index] = False #takes out from doors the one containing the car
goats = doors[mask]
```

```python
def host_choice(car,player):
  """
  car = door with the car
  player = door chosen by the player
  """
  if car == player:
    host_choice1 = np.array([np.random.choice(goats)])
    return host_choice1
  else:
    index_p = np.where(doors == player)[0][0]
    mask[[index_p,car_index]]= False
    host_choice2 = doors[mask]
    return host_choice2
```

```python
opened_door = host_choice(door_car, player_choice)
```

```python
#strategy one
def switch_doors(player,host):
  """
  player = player choice
  host = host choice
  """
  index_p = np.where(doors == player)[0][0]
  index_host = np.where(doors == host)[0][0]
  mask[[index_p,index_host]] = False
  final_choice = doors[mask]
  return final_choice
def stay_same(player):
  """
  player = player choice
  """
  return player
def choose_random(doors):
  return np.random.choice(doors)
```
