```python
import numpy as np
import matplotlib.pyplot as plt
```

This is simulation of the Monty Hall problem using vectorization techniques in which instead of sampling one trial at a time we sample N trials at once.

# Problem outline
* If the player's pick == car:
  * switching will mean he loses
  * staying means he wins
* If player's picks not car:
  * switching means he wins(since host is forced to open the door with a goat, leaving only as option the one with the car)
  * staying means he loses


```python
rng = np.random.default_rng() #generates random numbers
N = 10000000 #number of trials
car = rng.integers(0,3, size = N) # 1D array containing the locations of the car(0,1,2) at N trials
pick = rng.integers(0,3, size = N) # Player's pick
car2 = rng.integers(0,2, size = N) #locations of the car when only two doors are closed
pick2 = rng.integers(0,2, size = N)
```

```python
#we need to store the amount of times he wins with different strategies
stay_win = (pick == car) #stores the trials in which the player's initial pick was the car
switch_win = (pick!= car) #if he did not pick the car switching makes the player win
random_win = (pick2 == car2) #stores the number of times a player would win if he came after the host opened a door and chose randomly

stay_win_probability = stay_win.mean()
switch_win_probability = switch_win.mean()
random_win_probability = random_win.mean()
```

```python
probabilities = [stay_win_probability, switch_win_probability, random_win_probability]
labels = ['Stay', 'Switch', 'Random']

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, probabilities, edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}', ha='center', va='bottom')

plt.ylabel('Win Probability')
plt.title('Monty Hall Problem: Win Probabilities')
plt.ylim(0, 1)
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```
